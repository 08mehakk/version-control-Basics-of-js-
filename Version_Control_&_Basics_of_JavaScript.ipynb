Question 1.  Explain the purpose of version control systems like Git. Why are they essential in modern software development?

Answer -  Defination:- It is a system that records changes to a file over time.
   
  ###   Purpose of version control systems:-

   Version Control Systems like Git are tools that help track changes in files, rollback and collaboration too.
   1. Track changes:- Allows you to see what changes were made, who made them and when they were made.
   2. Rollback:- If a mistake is made or something goes wrong, version control lets you reverse.
   3. Collaboration:- Multiple people to work on the same project.
   4. Backup:- Provides a backup of your project, ensuring that your work is safe in case of data loss or other issues.

   ###   Essential in Modern Software Development:-
   1. Collaboration: Git enables multiple developers to work on the same project, reducing conflicts and errors.
   2. Flexibility: Git's distributed architecture allows developers to work offline and sync changes later.
   3. Scalability: Git can handle large projects with many contributors, making it suitable for complex software development.
   4. Security: Git's version control ensures that changes are tracked, and previous versions can be recovered if needed.
   5. Community: Git's popularity and widespread adoption mean that there are many resources available for learning and troubleshooting.

Question 2.  List the typical steps you would follow when working with Git to make changes to a project and share them on GitHub.

Answer - The following  steps to follow when working with Git to make changes to a project and share them on GitHub:

  Step 1: Clone the repository

   - git clone <repository-url>: Clone the repository from GitHub to your local machine.

 Step 2: Create a new branch

   - git branch <branch-name>: Create a new branch for your changes.
   - git checkout <branch-name>: Switch to the new branch.

 Step 3: Make changes

  - Edit files, add new files, or delete files as needed.

 Step 4: Stage changes

  - git add <file-name>: Stage specific files for commit.
  - git add .: Stage all changes in the repository.

  Step 5: Commit changes

  - git commit -m "<commit-message>": Commit changes with a meaningful message.

  Step 6: Pull latest changes

  - git pull origin <branch-name>: Pull the latest changes from the remote repository to ensure you have the most up-to-date code.

  Step 7: Resolve conflicts (if any)

  - If there are conflicts, resolve them manually and commit the resolved changes.

  Step 8: Push changes

  - git push origin <branch-name>: Push your committed changes to the remote repository on GitHub.

  Step 9: Create a pull request

  - Go to the GitHub repository and create a pull request to merge your branch into the main branch (e.g., main or master).

  Step 10: Review and merge

  - Wait for code review and approval.
  - Once approved, merge the pull request into the main branch.

Question 3. What is the purpose of a .gitignore file in a Git repository? Provide an example of how to ignore all .log files.

Answer - Purpose of .gitignore:-
  The .gitignore file specifies files or directories that Git should ignore in a repository. This helps to:

  1. Exclude unnecessary files: Prevents temporary or generated files from being tracked.
  2. Reduce repository size: Ignoring large files or directories can help keep the repository size manageable.
  3. Improve performance: Ignoring files can speed up Git operations like git status and git add.

#  Example: Ignoring all .log files

To ignore all .log files in your repository add the following line to .gitignore file:

- *.log

 This will tell Git to ignore any files with the .log extension.

 We can also ignore specific directories or files by adding their paths to the .gitignore file. For example:

-  logs/*.tmp

 This would ignore the logs/ directory and any files with the .tmp extension.
